```python
#KMEANS
from sklearn.cluster import KMeans
```

```python
distortions=[]
K=range(1,11)
for k in K:
    kmeanModel=KMeans(n_clusters=k)
    #fitting dataframe into the model
    kmeanModel.fit(encoder)
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions,'bx-',color='orange')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('Visualizing optimal k via Elbow Method')
```

```python
#Producing Silhouette Score for accuracy and choosing k-value
from sklearn.metrics import silhouette_score
range_of_clusters=list(range(3,10))
for n_clusters in range_of_clusters:
    clusters=KMeans(n_clusters=n_clusters)
    predictions=clusters.fit_predict(encoder)
    #Choosing centroids of clusters
    centers=clusters.cluster_centers_
    score=silhouette_score(encoder,predictions)
    print("For n_clusters={},silhouette score is {})".format(n_clusters,score))
```

```python
x_train, x_test_val,y_train, y_test=train_test_split(X,Y,test_size=0.25,random_state=1)
```

```python
# Feature Scaling
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test_val = scaler.transform(x_test_val)
```

```python
#AsyntheticSampling

asyn=ADASYN(sampling_strategy='minority',random_state=420,n_neighbors=10)
x_rs,y_rs=asyn.fit_sample(x_train,y_train)
```

```python
y_rd_df=pd.DataFrame(y_rs)
```

```python
y_rd_df['classification'].value_counts()
```

```python
#DecisionTree using ADASYN
decision_tree=DecisionTreeClassifier(criterion='entropy', max_depth=10)
```

```python
decision_tree.fit(x_rs,y_rs)
```

```python
y_pred=decision_tree.predict(x_test_val)
```

```python
print(pd.crosstab(y_test,y_pred,rownames=['Actual Observations'],
                  colnames=['Predicted Observations']))
print(classification_report(y_test,y_pred))
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
#RandomForestClassifier
random=RandomForestClassifier(criterion='entropy',
                              n_estimators=20,random_state=45)
random.fit(x_rs,y_rs)
```

```python
y_pred=random.predict(x_test_val)
```

```python
print(pd.crosstab(y_test,y_pred,rownames=['Actual Observations'],
                  colnames=['Predicted Observations']))
print(classification_report(y_test,y_pred))
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
# Import the classifiers
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import roc_curve, roc_auc_score

# Instantiate the classfiers and make a list
classifiers = [DecisionTreeClassifier(random_state=1234),
               RandomForestClassifier(random_state=1234),
              GradientBoostingClassifier(random_state=1234)]

# Define a result table as a DataFrame
result_table = pd.DataFrame(columns=['classifiers', 'fpr','tpr','auc'])

# Train the models and record the results
for cls in classifiers:
    model = cls.fit(x_rs, y_rs)
    yproba = model.predict_proba(x_test_val)[::,1]
    
    fpr, tpr, _ = roc_curve(y_test,  yproba)
    auc = roc_auc_score(y_test, yproba)
    
    result_table = result_table.append({'classifiers':cls.__class__.__name__,
                                        'fpr':fpr, 
                                        'tpr':tpr, 
                                        'auc':auc}, ignore_index=True)
    # Set name of the classifiers as index labels
result_table.set_index('classifiers', inplace=True)


fig = plt.figure(figsize=(8,6))

for i in result_table.index:
    plt.plot(result_table.loc[i]['fpr'], 
             result_table.loc[i]['tpr'], 
             label="{}, AUC={:.3f}".format(i, result_table.loc[i]['auc']))
    
plt.plot([0,1], [0,1], color='orange', linestyle='--')

plt.xticks(np.arange(0.0, 1.1, step=0.1))
plt.xlabel("Flase Positive Rate", fontsize=15)

plt.yticks(np.arange(0.0, 1.1, step=0.1))
plt.ylabel("True Positive Rate", fontsize=15)

plt.title('ROC Curve Analysis', fontweight='bold', fontsize=15)
plt.legend(prop={'size':13}, loc='lower right')

plt.show()

```
